```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from loss import contrastive_loss
```

```python
def predict_distance(model, image_a, image_b):
    a = np.expand_dims(image_a, 0)
    b = np.expand_dims(image_b, 0)
    return model.predict([a, b])
def show_pair_with_distance(image_a, image_b, distance):
    fig = plt.figure(figsize=(4., 4.))
    fig.suptitle(f'Distance: {distance[0][0]}', fontsize=16)
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(1, 2),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )

    for ax, idx in zip(grid, range(2)):
        ax.imshow(image_a if idx == 0 else image_b, cmap="gray")
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.text(distance, 0, 1)
    plt.show()
```

```python
model = tf.keras.models.load_model('output/contrastive_siamese_model.h5', compile=False)

cut_model = tf.keras.Sequential()
cut_model.add(model.layers[0])
cut_model.add(model.layers[2])

data = np.concatenate((np.load('output/data/train.npy'), np.load('output/data/test.npy')))
data_or = np.load('output/data/original.npy')
data_mod = np.load('output/data/modified.npy')
```

```python
for image_a, image_b in data[-20:]:
    dist = predict_distance(model, image_a, image_b)
    show_pair_with_distance(image_a, image_b, dist)
```
